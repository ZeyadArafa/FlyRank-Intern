## 1. My lane (or freestyle) and why

**Lane:** Refresh / Content Opportunity Scoring.

**Why:** The dataset contains a massive amount of content. Relying on simple rules like 'is it declining?' results in an unmanageable backlog for editors. We need a way to rank these pages so editors spend their limited capacity on the pages most likely to benefit from a refresh or most important to protect, rather than blindly working down an unranked list.

## 2. The question: decision, action, cost of a wrong call

- **Decision:** Which declining or underperforming page should an editor review and update first?
- **Action:** An editor reviews the flagged page and updates its content, title, or metadata.
- **Cost of a wrong call:** Wasted editor hours on a page that was experiencing natural seasonality or has low impact, and the missed opportunity cost from ignoring a high-value page actively losing rank.
- **Why ML helps:** A plain rule like `trend_direction == 'down'` flags thousands of pages. ML can learn the complex interaction between historical performance, age, current position, and competition to produce a priority score that beats a simple baseline.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')

total_pages = len(df)
declining_pages = len(df[df['trend_direction'] == 'down'])
declining_high_traffic = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] > 500)])

print(f"Total pages in starter dataset: {total_pages}")
print(f"Pages marked as declining ('down' trend): {declining_pages}")
print(f"Declining pages with significant traffic (>500 impressions/90d): {declining_high_traffic}")
print(f"\nThis shows that {declining_high_traffic} pages need attention—far too many for manual review without a ranked queue.")
```
